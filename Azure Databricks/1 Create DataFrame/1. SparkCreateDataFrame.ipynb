Spark Create Dataframe

```python
a = list(range(1, 20))
a
```

```python
type(a)
```

```python
spark.createDataFrame(a, 'int')  # converting normal python list of integers to spark data frame.
```

```python
b = [(x, x*2) for x in range(2,20)]
b
```

```python
spark.createDataFrame(b) #creating new dataframe from list of tuples
```

```python
spark.createDataFrame(b, 'index int, value int')  #createing dataframe using column name and data types.
```

```python
users_list = [(1, 'scott'), (3, 'donald'), (3, 'mickey'), (4, 'Elvis')]
users_list
```

```python
spark.createDataFrame(users_list)
```

```python
spark.createDataFrame(users_list, 'user_id int, user_first_name string')
```
